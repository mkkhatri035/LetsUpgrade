# Stop Words Removal

```python
stoppers = set(stopwords.words('english')) 
```

```python
def check_in_stoppers(word):
    if word in stoppers:
        return True

    return False
```

```python
def removestop(sentence):
    sentence=sentence.split()
    for word in sentence:
        if check_in_stoppers(word):
            sentence.remove(word)
    return sentence
```

```python
snow = nltk.stem.SnowballStemmer('english')
```

```python
temp =[]
for sentence in df.values:
    sentence = sentence.lower()     
    cleanr = re.compile('<.*?>')
    sentence = re.sub(cleanr, ' ', sentence)       
    sentence = re.sub(r'[?|!|\'|"|#|:|;]',r'',sentence)
    sentence = re.sub(r'[.|,|)|(|\|/]',r' ',sentence)
    sentence=removestop(sentence)
    sentence=[snow.stem(word) for word in sentence]
    temp.append(' '.join(sentence))
```

```python
protext=temp.copy()
```

```python
print(protext[0])
```

```python
processor = CountVectorizer(max_features=10000)
```

```python
processor.fit_transform(protext)
```
